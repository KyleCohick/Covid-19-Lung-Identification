## III. Data Preparation
Below is the code used to prepare the data to be fed into the Keras Classifier. The reductions mainly consisted of universalizing the images themselves. So, all pictures are taken in the posterioranterior position and resized to 244x244. The data was prepared using Keras's ImageDataGenerator class to organize all the images from the different files and conveniently carry their labels.

```python
from IPython.display import display
import itertools as it
from keras import optimizers
from keras.layers import Conv2D, Dropout, Flatten, MaxPooling2D
from keras.layers.core import Activation, Dense
from keras.models import Sequential
from keras.preprocessing import image
from keras.wrappers.scikit_learn import KerasClassifier
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import plotly
import plotly.graph_objects as go
import plotly.express as px
from PIL import Image
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
# creating instances of ImageDataGenerator for the training and validation imagesets
train_imageDataGenerator = image.ImageDataGenerator(rescale = 1./255, # to what size it's being rescaled
                                                    shear_range = 0.2, # shear intensity
                                                    zoom_range = 0.2) # how much to zoom in

val_imageDataGenerator = image.ImageDataGenerator(rescale=1./255)

# reshaping the test and validation images
train_generator = train_imageDataGenerator.flow_from_directory(
    'data/prepped_datasets/training', # directory containing the images
    target_size = (224,224), # the dimensions to which all images found will be resized
    batch_size = 32, # size of batches
    class_mode = 'binary', # determines type of label array returned, since there are two classes, binary is fine
    shuffle=True) # shuffles order images are interred
validation_generator = val_imageDataGenerator.flow_from_directory(
    'data/prepped_datasets/val',
    target_size = (224,224),
    batch_size = 32,
    class_mode = 'binary', 
    shuffle=True)

# sources:
# https://keras.io/preprocessing/image/
# https://blog.keras.io/building-powerful-image-classification-models-using-very-little-data.html
# https://towardsdatascience.com/detecting-covid-19-using-deep-learning-262956b6f981
```

## IV. Modeling
The models were first created based off of a 3-layer Convolutional Neural Network created using Keras's Conv2D layers. This 3-layer model was compared then against a 5-layer and 6-layer model design all using similary hyperparamters to see which faired better on the test data. The measurements used to differenciate the model were validation loss, validation accuracy, loss, and accuracy. Additionally, the models were then tested on the validation data and the results were shown using classification reports, confusion matrixes, and heatmaps. It was determined that the 3-layer model was giving the most consistent results, so the 3-layer model was chosen for further hyperparameter tuning. The hyperparameters worked with are described in this dictionary:
param_grid = {
    'optimizer' : ['adam', 'sgd', 'rmsprop'],
    'activation' : ['relu', 'tanh'],
    'epochs' : [5, 10, 15],
    'epoch_steps' : [8, 16, 32]
}
However, because of my use of Keras's ImageDataGenerator, I could not use scikit-learn's GridSearchCV method as it only works with 1-dimensional data. So, I performed the task of GridSearchCV through my own code and compared the results. The full code and process is available in the Modeling.ipynb file in this repository. The hyperparameters that gave the highest accuracy and the lowest loss were (relu, 32, 15, adam). Below is the code for the final model.

```python
# model creation
final_model = Sequential()

# input layer
final_model.add(Conv2D(32, (3, 3), input_shape=(224, 224, 3)))
final_model.add(Activation('relu'))
final_model.add(MaxPooling2D(pool_size=(2, 2)))

# hidden layer 1
final_model.add(Conv2D(32, (3, 3)))
final_model.add(Activation('relu'))
final_model.add(MaxPooling2D(pool_size=(2, 2)))

# hidden layer 2
final_model.add(Conv2D(64, (3, 3)))
final_model.add(Activation('relu'))
final_model.add(MaxPooling2D(pool_size=(2, 2)))

# output layer
final_model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors
final_model.add(Dense(64))
final_model.add(Activation('relu'))
final_model.add(Dropout(0.5))
final_model.add(Dense(1))
final_model.add(Activation('sigmoid'))

final_model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

final_model_result = final_model.fit_generator(
    train_generator,
    steps_per_epoch=15,
    epochs=32,
    validation_data=validation_generator,
    validation_steps=2,
    verbose=1)
```

## V. Performance Results
With the model trained, we can explore how the model performed on the validation data we have randomly separated from the full dataset. The split can be seen within the data/prepped_datasets folder within the repository. The split was 120/30 Normal images and 114/30 Covid images. Below is the code to get predictions from the model and compare them against the true values.

```python
y_true = []
y_pred = []

# get predictions for all in the Normal validation set
for file in os.listdir("./data/prepped_datasets/val/Normal/"):
    img=image.load_img("./data/prepped_datasets/val/Normal/" + file, target_size=(224,224))
    img=image.img_to_array(img)
    img=np.expand_dims(img, axis=0)
    pred=final_model.predict_classes(img)
    y_pred.append(pred[0,0])
    y_true.append(1)
    
for file in os.listdir("./data/prepped_datasets/val/Covid/"):
    img=image.load_img("./data/prepped_datasets/val/Covid/" + file,target_size=(224,224))
    img=image.img_to_array(img)
    img=np.expand_dims(img,axis=0)
    pred=final_model.predict_classes(img)
    y_pred.append(pred[0,0])
    y_true.append(0)
    
y_true=np.array(y_true)
y_pred=np.array(y_pred)
```

### Confusion Matrix

```python
confusion_matrix(y_true, y_pred)
```

```python
sns.heatmap(confusion_matrix(y_true, y_pred), annot=True)
```

### Classification Report

```python
print(classification_report(y_true, y_pred))
```
