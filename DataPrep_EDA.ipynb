## Data Preparation
This code below will randomly build the dataset of normal images from among the 1,300 images in the normal dataset. I set it to only choose 150.

```python
# this script is heavily inspired by Adrian Rosebrock's script written to do a similar process: 
# https://www.pyimagesearch.com/2020/03/16/detecting-covid-19-in-x-ray-images-with-keras-tensorflow-and-deep-learning/
# import the necessary packages
from imutils import paths
import random
import shutil
import os

# path to pneumonia directory
SOURCE_DIR = r"data\chest_xray\train\NORMAL"
# path to output directory
OUTPUT_DIR = r"data\prepped_datasets\training\normal"
# number of images to take from the dataset
IMAGE_NUM = 150

# grab all training image paths from the Kaggle X-ray dataset
image_paths = list(paths.list_images(SOURCE_DIR))

# shuffle images and choose the 0th through IMAGE_NUM images
random.shuffle(image_paths)
image_paths = image_paths[:IMAGE_NUM]

# loop over the image paths
for (i, image_path) in enumerate(image_paths):
    # extract the filename from the image path and then construct the path to the copied image file
    file_name = image_path.split(os.path.sep)[-1]
    output_path = os.path.sep.join([OUTPUT_DIR, file_name])

    # copy the image
    shutil.copy2(image_path, output_path)
```

Then we must take the training data and choose some of it to be used for validation. This will be done randomly and use the script above and only about 30 of each picture will be used. This makes for roughly a 70%/30% split for training and testing data.

```python
from imutils import paths
import random
import shutil
import os

# path to pneumonia directory
SOURCE_DIR = r"data\prepped_datasets\training\Covid"
# path to output directory
OUTPUT_DIR = r"data\prepped_datasets\val\Covid"
# number of images to take from the dataset
IMAGE_NUM = 30

# grab all training image paths from the Kaggle X-ray dataset
image_paths = list(paths.list_images(SOURCE_DIR))

# randomly sample the image paths
random.shuffle(image_paths)
image_paths = image_paths[:IMAGE_NUM]

# loop over the image paths
for (i, image_path) in enumerate(image_paths):
    # extract the filename from the image path and then construct the path to the copied image file
    file_name = image_path.split(os.path.sep)[-1]
    output_path = os.path.sep.join([OUTPUT_DIR, file_name])

    # copy the image
    shutil.copy2(image_path, output_path)
    
    # deletes the original image
    input_path = os.path.sep.join([SOURCE_DIR, file_name])
    if os.path.isfile(input_path): # if the file exists, delete it
        os.remove(input_path)
```

The data right now is uniform in view, but not in a format that is convenient for the algorithm to analyze. In order to make the pictures as uniform as possible, the code below will be used to resize the pictures and process them into the Keras algorithms. 

```python
import numpy as np
import keras
from keras.preprocessing import image

# creating instances of ImageDataGenerator for the training and validation imagesets
train_imageDataGenerator = image.ImageDataGenerator(rescale = 1./255, # to what size it's being rescaled
                                                    shear_range = 0.2, # shear intensity
                                                    zoom_range = 0.2) # how much to zoom in

val_imageDataGenerator = image.ImageDataGenerator(rescale=1./255)

# reshaping the test and validation images
train_generator = train_imageDataGenerator.flow_from_directory(
    'data/prepped_datasets/training', # directory containing the images
    target_size = (224,224), # the dimensions to which all images found will be resized
    batch_size = 32, # size of batches
    class_mode = 'binary', # determines type of label array returned, since there are two classes, binary is fine
    shuffle=True) # shuffles order images are interred
validation_generator = val_imageDataGenerator.flow_from_directory(
    'data/prepped_datasets/val',
    target_size = (224,224),
    batch_size = 32,
    class_mode = 'binary', 
    shuffle=True)

# sources:
# https://keras.io/preprocessing/image/
# https://blog.keras.io/building-powerful-image-classification-models-using-very-little-data.html
# https://towardsdatascience.com/detecting-covid-19-using-deep-learning-262956b6f981
```
